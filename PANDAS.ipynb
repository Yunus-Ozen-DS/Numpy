```python
data={'x':['al','vl','sl','al','vl','sl'],
      'b':['11','22','33','11','22','33'],
      'cu':['mm','nn','uu','tt','yy','oo'],
      'W':[1,1,5,5,88,2],
      'Q':[8,6,5,4,2,1]}
df=pd.DataFrame(data)
df
```

```python
df.groupby('x').mean()
```

```python
df.groupby(["x","b","cu"])[["W","Q"]].agg('mean').stack()
```

```python
df.groupby(["x","b","cu"])[["W","Q"]].agg('mean').unstack().unstack()
```

```python
df.unstack().unstack().unstack().unstack()
```
